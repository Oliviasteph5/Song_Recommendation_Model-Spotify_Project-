```python
#define a function to get the playlist ids/songs using the Spotify Web API
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
#retrieves the rock playlist from spotify
rock = get_playlist_tracks('Spotify','37i9dQZF1DWXRqgorJj26U')

#create the ids of rock songs
rock_ids =[]
for i in rock:
    rock_ids.append(i['track']['id'])
    
#check the length
len(rock_ids)
```

```python
#displays the extended list of list of rock_af
rock_af = []
for i in rock_ids_chopped:
    rock_af.extend(sp.audio_features(i))
```

## choosing k

```python
K = range(2, 40)
inertia = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(modplaylist_df_scaled_df)
    inertia.append(kmeans.inertia_)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
#using silhuoette

K = range(2, 40)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234, n_init='auto')
    kmeans.fit(modplaylist_df_scaled_df)
    
    filename = "kmeans_" + str(k) + ".pickle"
    # with open(filename, "wb") as f:
    #     pickle.dump(kmeans,f)
    
    silhouette.append(silhouette_score(modplaylist_df_scaled_df, kmeans.predict(modplaylist_df_scaled_df)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette Method showing the optimal k')
```

## My Song Recommender

```python


def get_song_id(user_input):
    result = sp.search(q=user_input,limit=1)       # put user input into spotify api to get song information 
    song_id = result['tracks']['items'][0]['id']   # extract song_id from song information
    return song_id
                         

def song_display(song):                                                 # function to display an IFrame with the song -> function uses song_id as parameter
    display(IFrame(src="https://open.spotify.com/embed/track/"+ song,            
            width="320",                                                                   
            height="80",
            frameborder="0",
            allowtransparency="true",
            allow="encrypted-media",
            ))
    

def recommender(user_input):
    # get the id of the song via api
    api_results = sp.search(q=user_input,limit=1)     # This returns a dictionary
    song_id = api_results['tracks']['items'][0]['id'] # this gives the id
    
    # Song was not found in the playlist_DF -->
    features = sp.audio_features(song_id)[0]                                                        # Spotify Api call to get the audio features of the requested song (dict)
    relevant_features = {k:v for k,v in features.items() if k in modplaylist_df_scaled_df.keys()}   # create a new Dict with only the relevant features for the clustering
    df = pd.DataFrame([relevant_features])                                                          # create a DF of the features (only one row)
    scal = scaler.transform(df)                                                                     # transform the DF with the scaler
    prediction = kmeans.predict(scal)[0]                                                            # use Kmeans to predict the cluster #kmeans.predict(modplaylist_df_scaled_df
    print('Predicted Cluster of User_Input Song: '+str(prediction))                                 # DEBUG: Check the predicted cluster
    cluster_df = playlist_df[playlist_df.cluster == prediction]                                     # create a new DF with only songs from the cluster
    cluster_df = cluster_df.reset_index(drop = True)                                                # reset index
    random_song_index = random.randint(0, len(cluster_df)-1)                                        # generate a random integer between 0 and the number of songs in the new df
    random_song_id = cluster_df['id'][random_song_index]                                            # randomly pick a song from the cluster
    print('Recommended Song ID: '+random_song_id)                                                   # DEBUG: Check the ID of the recommended song
       
    #print(f'Good Choice, here is your Song. You might also enjoy {random_track_title} by {random_track_artist}') # print                      
        
    song_display(random_song_id)
    
def mySong_recommender():
    user_input = input("Enter your current hot song: ")
    
    song_id = get_song_id(user_input)   # get song_id using the func above
    song_display(song_id)               # pLay song using the song_display function above

    # Convert the "SONG" column values to lowercase for case-insensitive comparison
    hot100_df_songs = hot100_df["SONG"].str.lower()
    fav_df_songs = fav_df["SONG"].str.lower()

    #if user_input is in hot100_df_songs, play recommended song
    if user_input in hot100_df_songs.values:
        # Recommend another song from hot100_df
        rec_song_hot = hot100_df[hot100_df_songs != user_input]["SONG"].iloc[0]           # copy the hot100_df row by row except for the row that matches the user_input and recommend the first song
        print(f"Great choice! You might also like '{rec_song_hot}' from the hot list.")   # print
        
        hot_song_id = get_song_id(rec_song_hot)   # get hot_song_id using the func above
        song_display(hot_song_id)                 # pLay hot_song_id using the song_display function above

         # Check for recommendation in fav_df as well
        if user_input in fav_df_songs.values:
            # Recommend another favorite song different from hot100_df recommendation
            rec_song_fav_df = fav_df[fav_df_songs != user_input]      # copy the fav_df row by row except for the row that matches the user_input
            random_song_index = random.randint(0, len(rec_song_fav_df)-1)    # get a random number between 0 and the number of songs in the rec_song_fav_df
            rec_song_fav = rec_song_fav_df["SONG"].iloc[random_song_index]  # get the Song Column from the rec_song_fav_df, pick a random song using the random index
            
            while rec_song_fav == rec_song_hot:                                # Ensure the recommendation is different
                random_song_index = random.randint(0, len(rec_song_fav_df)-1)    # get a random number between 0 and the number of songs in the rec_song_fav_df
                rec_song_fav = rec_song_fav_df["SONG"].iloc[random_song_index]      # get the Song Column from the rec_song_fav_df, pick a random song using the random index
            print(f"Awesome taste! You might also like '{rec_song_fav}' from my favorite songs.")  #print and recommend a different song
            
            fav_song_id = get_song_id(rec_song_fav)       # get fav_song_id using the func above
            song_display(fav_song_id)                     # pLay song using the song_display function above
    else:
        print(f'Sorry, we could not find "{user_input}" in the hot100 songs ') #otherwise print when song not found in hot100 songs
        
        if user_input in fav_df_songs.values:
            # Recommend another favorite song
            rec_song_fav_df = fav_df[fav_df_songs != user_input]                         # copy the fav_df row by row except for the row that matches the user_input
            random_song_index = random.randint(0, len(rec_song_fav_df)-1)                  # get a random number between 0 and the number of songs in the rec_song_fav_df
            recommended_song = rec_song_fav_df["SONG"].iloc[random_song_index]           # get the Song Column from the rec_song_fav_df, pick a random song using the random index
            print(f"but you might also like '{recommended_song}' from my favorite songs.")
         
            song_id2 = get_song_id(recommended_song)      # get song_id2 of recommended song using the func above
            song_display(song_id2)                        # pLay song using the song_display function above

        else:
            print(f"Sorry, '{user_input}' is not in the hot list or my favorite songs.") #print
            print(f'mySong_recommender: Great Choice, here is your Song. You might also enjoy:')             #print
            recommender(user_input) #displays the songs
```
